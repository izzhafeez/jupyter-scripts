```python
def run(n):
    points = list(zip(np.random.uniform(0,1,n),np.random.uniform(0,1,n)))
    mydict = {n:0 for n in range(n)}
    for i in Telegram(n):
        minsrc = 100000
        mindist = 1000
        for j in range(n):
            if i != j:
                distance = dist(points[i],points[j])
                if mindist > distance:
                    mindist = distance
                    minsrc = j
        mydict[minsrc] += 1
    return max(mydict.items(),key=operator.itemgetter(1))
```

```python
run(100000)
```
