```python
from wilcoxon import *
import requests
import re
from bs4 import BeautifulSoup
import json
import os
from tqdm import tqdm
import pandas as pd
import ast
import collections
```

```python
def get_link(title,artist=''):
    page_number = 1
    while True:
        try:
            if page_number > 10:
                return ['']
                break
            link = 'https://www.ultimate-guitar.com/search.php?search_type=title&value=' + title.lower() + '&page=' + str(page_number)
            artist_sep = artist.lower().split()
            site = requests.get(link).text
            links = re.findall('(https://tabs.ultimate-guitar.com/tab/.+?)(?=&quot;)',site)
            results = string.findAllWords(links,yes=artist_sep,no=['guitar-pro','official'])
            if len(results) > 0:
                return results
                break
            else: page_number += 1
        except: return ['']
                
```

```python
def getChords(song,artist=''):
    try:
        link = get_link(song,artist)[0]
        site = BeautifulSoup(requests.get(link).text,'html.parser')
        content = json.loads(site(class_='js-store')[0]['data-content'])['store']['page']['data']['tab_view']['wiki_tab']['content']
        sections = {}
        for section in re.split(r'(?=(\[|\()[A-Z]+.*?(\]|\)))',content):
            sections[section.split(']')[0][1:]] = re.findall(r'\[ch\](.*?)\[/ch\]',section)
        return [sections,link]
    except: return []
```

```python
itunesTxt = open('Documents/Data/itunes.txt').read()
rows = []
for row in itunesTxt.split('\n'):
    rows.append(row.split('\t'))
```

```python
newRows = []
for row in tqdm(rows):
    song = re.sub('[^A-Za-z ]',' ',row[0].split('(')[0]).strip()
    artist = re.sub('[^A-Za-z ]',' ',row[2])
    row.extend(getChords(song,artist))
    newRows.append(row)
```

```python
alltypes = []
allSections = []
for chordSet in allChords:
    sections = {'Verse':[],'Chorus':[],'Pre':[],'Outro':[],'Instrumental':[],'Intro':[]}
    chordSet = ast.literal_eval(chordSet)
    for chordDictKey in chordSet:
        for sectionKey in sections:
            if sectionKey in chordDictKey:
                progression = chordSet[chordDictKey]
                try:
                    if len(progression) % 7 == 0:
                        phrase = [', '.join(progression[i:i+7]) for i in range(0, len(progression), 7)]
                        phrase = list(set(phrase))
                    elif len(progression) % 4 == 0:
                        phrase = [', '.join(progression[i:i+4]) for i in range(0, len(progression), 4)]
                        phrase = list(set(phrase))
                    elif len(progression) % 3 == 0:
                        phrase = [', '.join(progression[i:i+3]) for i in range(0, len(progression), 3)]
                        phrase = list(set(phrase))
                    else: phrase = progression
                except Exception as e: 
                    print(e)
                    phrase = progression
                sections[sectionKey].extend(phrase)
    allSections.append(sections)
                
```
