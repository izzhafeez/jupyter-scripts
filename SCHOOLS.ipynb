```python
schools = sheets.getGeo("Schools")
```

```python
names = schools.Name.str.replace("St. ","St ").str.strip()
```

```python
result = []
for name in tqdm(names):
    res = latlong(name)
    if res == (0,0):
        res = latlong(re.sub("\(Primary\)|\(Secondary\)","",name).strip())
    result.append(res)
```

```python
## Finding year

secondary = spiderman.website("https://en.wikipedia.org/wiki/List_of_secondary_schools_in_Singapore")

def getYear(link):
    
    school = spiderman.website(link)
    
    for table in school.html("table"):

        if "Established" in table.text:

            return int(re.findall("\d{4}", table("th", text = re.compile("Established"))[0].find_next("td").text)[0])

L = []

for row in tqdm(secondary.html("table")[0]("tr")):
    
    rowCells = row("td")
    
    if len(rowCells) > 0:
        
        try:
            
            result = [rowCells[0].text, getYear("https://en.wikipedia.org" + rowCells[0]("a")[0]["href"])]
            print(result)
            L.append(result)
        
        except: None
```

```python
## Updating values in sheets

schools = sheets.getGeo("Schools")
schools.merge(pd.DataFrame(L), left_on = "Name", right_on = 0, how = "left")[1].to_clipboard(index = False, header = False)
set([x[0] for x in L]) - set(schools.Name.tolist())
```
