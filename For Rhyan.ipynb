```python
individual = rhyanSheet.find('Individual Kanji').to_frame().dropna(how='all').fillna('')
readings = rhyanSheet.find('Readings').to_frame().dropna(how='all').reset_index().iloc[:,:4].fillna('')
readings.columns = ['Kanji','Reading','Meaning','Category']
individual.columns = ['かんじ','くんよみ','おんよみ','いみ','Lesson','Duplicate','Dupe']
```

```python
individual.iloc[0:len(individual)]
```

```python
def vocab(start=0,end=len(individual),number=1):
    
    samplingResult = individual.iloc[start:end].sample(number).iloc[0]
    kan,kun,on,imi = samplingResult[['かんじ','くんよみ','おんよみ','いみ']]

    text = f'Word: {kan}\n\nかんじ: {kan}\nくんよみ: {kun}\nおんよみ: {on}\nいみ: {imi}\n\n\nOther words: \n\n'
    includesIt = readings.loc[readings['Kanji'].apply(lambda x: kan in x)].to_dict('records')
    
    for word in includesIt:
        text += '\n'.join([word['Kanji'],word['Reading'],word['Meaning']]) + '\n\n'

    telegram.sendRhyan(text)
```

```python
import time
while(True):
    time.sleep(120)
    vocab()
```
