```python
def urlFinder(x):    
    try:
        site = spiderman.website(x)
        site.getLists()
        text = site.site.find_all(class_='entry-content')[0].getText()
        splitted = re.split('Parameters :',text.replace('About : ','').replace('\n',''))
        functionName = re.findall('^(.*?)\(',splitted[0])[0]
        purpose = re.findall(': (.*)$',splitted[0])[0]
        params = re.findall('\r *([^\r]*?) *:',splitted[1])
        attributes = re.findall(': *([^\r]*?)\r',splitted[1])
        return [purpose,list(zip(params,attributes))]
    except: return ['',[('','')]]
```

```python
df['purposeParams'] = df['1'].apply(lambda x: urlFinder(x))
df['purpose'] = df['purposeParams'].apply(lambda x: x[0])
df['params'] = df['purposeParams'].apply(lambda x: x[1])
df = df.explode('params')
df = df.fillna(['',''])
df['attributes'] = df['params'].apply(lambda x: x[1])
df['others'] = df['params'].apply(lambda x: x[0])
```
