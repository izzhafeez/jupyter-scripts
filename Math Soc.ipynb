```python
seven = [i*7 for i in range(100)]
eleven = [i*11 for i in range(100)]
thirteen = [i*13 for i in range(100)]
```

```python
import itertools
```

```python
s1 = list((i, abs(i[1] - i[0])) for i in itertools.product(seven, eleven))
s2 = list((i, abs(i[1] - i[0])) for i in itertools.product(thirteen, eleven))
s3 = list((i, abs(i[1] - i[0])) for i in itertools.product(seven, thirteen))
```

```python
s = [*s2]
```

```python
mins = {}
for i in s:
    if i[1] not in mins:
        mins[i[1]] = min(i[0])
    else:
        mins[i[1]] = min(mins[i[1]], *i[0])
```

```python
for i in [(k, mins[k]) for k in sorted(mins,key=lambda x: mins[x])]:
    print(i)
```
