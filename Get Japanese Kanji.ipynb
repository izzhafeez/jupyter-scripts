```python
site = spiderman.website('https://kanji.jitenon.jp/cat/kyoiku.html')
```

```python
df = pd.Series()
for i in site.lists:
    if re.compile('^.\n').search(i[0]):
        df = pd.concat([df,pd.Series(i)])
```

```python
site = spiderman.website('https://kanji.jitenon.jp/kanji/001.html')
```

```python
for wordSoup in site.site(class_='word'):
    word = wordSoup.get_text()
    yomi = ''
    sentence = ''
    wordSoup = wordSoup.parent
    try: yomi = wordSoup(class_='yomi')[0].get_text()
    except: continue
    try: sentence = wordSoup(class_='p')[0].get_text()
    except: continue
    print(word,yomi,sentence)
```
