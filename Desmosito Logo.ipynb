```python
from PIL import Image

img = Image.open('../Website 3.0/img/desmosito.png')
img = img.convert("RGBA")
datas = img.getdata()

newData = []
for item in datas:
    if item[0] < 100 and item[1] < 100 and item[2] < 100:
        newData.append((0, 0, 0, 0))
    else:
        newData.append((255, 255, 255, 255))

img.putdata(newData)
img.save("../Website 3.0/img/desmosito bw.png", "PNG")
```
