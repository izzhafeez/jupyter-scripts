```python
import pandas as pd
import urllib
import ssl
import re
import time
from tqdm import tqdm
from hanziconv import HanziConv
import string
ssl._create_default_https_context = ssl._create_unverified_context
```

```python
df = pd.read_csv('Documents/Vocab.csv')
```

```python
chinese_chars_used = df[df['Language'] == '中文'].dropna(thresh=2)['Word'].apply(lambda x: list(x)).explode().unique().tolist()
```

```python
raw_request = 'http://bcc.blcu.edu.cn/zh/search/0/%E6%8A%B5%E8%BE%BE'
def old_searcher(word,trans):
    time.sleep(1)
    url = raw_request+trans
    req = urllib.request.Request(url)
    with urllib.request.urlopen(req) as response:
        the_page = response.read().decode('utf-8')
    outcomes = re.findall(r'>(\S{1,8})</a></td>\n\t\t\t<td class="nowrap">(.{1,8})</td>\n\t\t\t<td>(.*?)\n',str(the_page))
    print(outcomes)
    valid_outcome = [i for i in outcomes if word == i[1]][0]
```

```python
raw_request = 'https://koreanhanja.app/'
def english_searcher(english,master_df=master_df):
#     time.sleep(1)
    url = raw_request+english
    req = urllib.request.Request(url)
    with urllib.request.urlopen(req) as response:
        the_page = response.read().decode('utf-8')
    outcomes = re.findall(r'>(\S{1,8})</a></td>\n\t\t\t<td class="nowrap">(.{1,8})</td>\n\t\t\t<td>(.*?)\n',str(the_page))
    o = pd.DataFrame(outcomes,columns=['Chinese','Korean','Usages'])
    o[0] = o[0].apply(HanziConv.toSimplified)
    return(o[o[0].apply(lambda x: len(set(list(x)) & set(chinese_chars_used)) == len(x))])
```

```python
master_df = pd.DataFrame(columns=['Chinese','Korean','Usages'])
for char in set(string.ascii_lowercase):
    master_df = pd.concat([master_df,english_searcher(char)])
```
