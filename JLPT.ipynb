```python
def tryJlptSensei(x):
    try:
        site = spiderman.website(x)
        example = site.site(class_='example-main')[-1]
        romaji = example.find_next(class_='alert-info').get_text()
        translation = example.find_next(class_='alert-primary').get_text()
        return [example.get_text(),romaji,translation]
    except: return ['','','']
```

```python
vocab.getSheets()
japanese = vocab.getVocab('Japanese')
```

```python
links = japanese['Link'].to_list()
```

```python
Ls = []
for link in tqdm(links):
    Ls.append(tryJlptSensei(link))
```

```python
Ls
```

```python
others = japanese[['Examples','English','Translation']].to_numpy()
```

```python
newLs = []
for L in range(len(Ls)):
    if Ls[L][0] == '':
        newLs.append(others[L])
    else: newLs.append(Ls[L])
```
