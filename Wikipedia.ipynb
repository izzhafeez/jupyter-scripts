```python
def spiderman(keywords):
    ps = []
    hrefs = []
    base_wiki = 'https://en.wikipedia.org/wiki/'
    for keyword in keywords:
        keyword = '_'.join(keyword.split())
        iterations = 0
        site = website(base_wiki+keyword)
        site.relevant_hrefs(keywords)
        site.relevant_ps(keywords)
        hrefs.extend(site.relevant_href)
        ps.extend(site.relevant_p)
        for href in hrefs:
            if iterations < 100:
                site = website(base_wiki+keyword)
                site.relevant_hrefs(keywords)
                site.relevant_ps(keywords)
                hrefs.extend(site.relevant_href)
                ps.extend(site.relevant_p)
                iterations += 1
    return ps,hrefs
```

```python
p,href = spiderman(['donald trump','president trump'])
```
