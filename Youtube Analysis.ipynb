```python
import pandas as pd
import numpy as np
import re
```

```python
df = pd.read_csv('/Users/izzhafeez/Documents/youtube channels.csv')
```

```python
df = df[df['Tier'].notnull()]
```

```python
def clean_vid(x):
    try: return int(str(x).replace(',',''))
    except: return int(x)
df['Videos'] = df['Videos'].apply(clean_vid)
```

```python
def clean_subs(x):
    if 'M' in x:
        return int(float(x.replace('M',''))*1000000)
    elif 'K' in x:
        return int(float(x.replace('K',''))*1000)
    else: return int(x)
df['Subs'] = df['Subs'].apply(clean_subs)
```

```python
df['Tier Num'] = df['Tier'].replace({'MINE':0,'GOD':1,'SSS':2,'SS':3,'S':4,'AS':5,'A':6,'B':7,'C':8,'D':9,'E':10,'F':11})
df['Tier Score'] = 3 * df['Tier Num'] + df['Sub-tier']
df['Tier Score'] = df['Tier Score'].apply(lambda x: int(x))
```

```python
df['Total Category'] = df['Category'] + ' ' + df['Sub-category']
pd.pivot_table(df, index = ['Total Category'], values = ['Tier Score','Still Watch'], aggfunc = {'Tier Score': np.mean,'Still Watch': 'count'}).sort_values(by = 'Still Watch').to_csv('/Users/izzhafeez/Documents/Youtube/youtube results.csv')
```
