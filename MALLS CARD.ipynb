```python
def bestStat(stat):
    if stat > 85:
        return "#3C8F41"
    elif stat < 16:
        return "#C10105"
    else: return "#000000"

def bestestStat(stat):
    if stat > 95 or stat < 6:
        return '<span style="color:#EFD065;">â˜… </span>'
    else: return ""
    
def comma_adder(num):
    return '{:,}'.format(num)
```

```python
def merge_images(files, n):
    
    images = [Image.open(x) for x in files]

    (width, height) = images[0].size

    # result = Image.new('RGB', (n * width, n * height))
    result = Image.new('RGB', (int(2480), int(3508)), color="white")
    
    for index, image in enumerate(images):
        result.paste(image, (width * (index % n), height * math.floor(index / n)))
    
    return result
```

```python
for i in range(12):
    merge_images(["Malls/MALLS CLEAN CARDS/" + x + ".png" for x in malls["Name"].tolist()]
                  [i*9:(i+1)*9], 3).save("Malls/MALLS FINAL/Set" + str(i) + ".png")
```
