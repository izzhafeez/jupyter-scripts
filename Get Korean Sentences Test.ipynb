```python
vocab.getSheets()
korean = vocab.getVocab('Korean')
```

```python
korean['Chinese'] = korean['Word'].apply(string.tryHanja)
korean['Pinyin'] = korean['Chinese'].apply(string.tryPinyin)
```

```python
words = korean['Word'].tolist()
```

```python
chinese = []
for word in tqdm(words):
    chinese.append(string.tryHanja(word))
```

```python
pd.Series(chinese).to_clipboard()
```

```python
pinyin = []
for word in chinese:
    pinyin.append(string.tryPinyin(word))
```
